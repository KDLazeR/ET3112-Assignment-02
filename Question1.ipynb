```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# 1. Load the cropped image
# Note: Ensure the image file is in the same folder as this notebook, 
# or update the path below to match where you saved it.
img = cv.imread('1c.jpg', cv.IMREAD_GRAYSCALE)

# 2. Apply Canny Edge Detector
# The assignment specifies minVal=550 and maxVal=690 [cite: 882]
edges = cv.Canny(img, 550, 690) [cite: 888]

# 3. Plot both the original image and the edge image [cite: 884]
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Cropped Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detector Output')
plt.axis('off')

plt.show()

# 4. Extract the x and y coordinates of the edges [cite: 887]
indices = np.where(edges != 0) [cite: 889]
x = indices[1] [cite: 892]
y = indices[0] [cite: 893]

print(f"Extracted {len(x)} edge points.")
```
