```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Re-load the image and extract edges for this standalone notebook
image_filename = 'IMages/1c.jpg'
img = cv.imread(image_filename, cv.IMREAD_GRAYSCALE)

if img is None:
    print(f"Error: Could not find {image_filename}. Check your folder path!")
else:
    # Apply Canny and get coordinates
    edges = cv.Canny(img, 550, 690)
    indices = np.where(edges != 0)
    x = indices[1]
    y = indices[0]

    # Least-Squares Fit Calculation
    # We want to fit a line y = mx + c. 
    # We set up a matrix A with the x values and a column of ones.
    A = np.vstack([x, np.ones(len(x))]).T
    
    # Solve for m (slope) and c (intercept) using numpy's least squares solver
    m, c = np.linalg.lstsq(A, y, rcond=None)[0]
    
    print(f"Estimated Slope (m): {m:.4f}")
    print(f"Estimated Intercept (c): {c:.4f}")

    # Calculate the y values for the fitted line
    y_fit = m * x + c

    # Create the Scatter Plot with the Fitted Line
    plt.figure(figsize=(8, 8))
    
    # Plot original edge points
    plt.scatter(x, y, s=2, c='blue', label='Edge Points') 
    
    # Plot the least-squares line
    plt.plot(x, y_fit, color='red', linewidth=2, label=f'Least Squares Fit Line')

    # Invert the y-axis so the plot matches the orientation of the original image
    plt.gca().invert_yaxis() 

    plt.title('Least-Squares Fit of Crop Field Edges')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6) 
    
    plt.show()
```
